# filter

```python
type(True)
```

```python
booleans =[]
for lenght in movies.duration:
    if lenght >=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long].head()
```

```python
is_long=movies.duration>=200
is_long.head()
```
