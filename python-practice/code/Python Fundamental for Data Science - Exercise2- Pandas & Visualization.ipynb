<b> Problem 4 </b>

Plot the distribution of all numerical features and the categorical target using matplotlib and observe the plots.

```python
fig, ax=plt.subplots(2,2, figsize=(20,8))
indexes_l=[0,0,1,1]
indexes_r=[0,1,0,1]

for ind, col in enumerate(iris.columns):
    if col!='species':
        data=iris[col]
        sns.distplot(data, kde=True, rug=True, ax=ax[indexes_l[ind]][indexes_r[ind]])
    
```

```python
sns.catplot(x="species", y='sepal_length', data=iris,  jitter=False)
```

```python
fig, ax=plt.subplots(2,2, figsize=(20,8))
indexes_l=[0,0,1,1]
indexes_r=[0,1,0,1]

for ind, col in enumerate(iris.columns):
    if col!='species':
        sns.boxplot(x="species", y=col, data=iris,  ax=ax[indexes_l[ind]][indexes_r[ind]])
```

```python
fig, ax=plt.subplots(2,2, figsize=(20,8))
indexes_l=[0,0,1,1]
indexes_r=[0,1,0,1]

for ind, col in enumerate(iris.columns):
    if col!='species':
        sns.swarmplot(x="species", y=col, data=iris,  ax=ax[indexes_l[ind]][indexes_r[ind]])
```
